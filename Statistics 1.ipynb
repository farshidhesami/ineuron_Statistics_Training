## What is Statistics?
#### Statistics is the study of the collection, analysis, interpretation, presentation, and organization of data. 

## What is Data? 
#### Data is a collection of facts, such as numbers, words, measurements, observations or just descriptions of things. such as height, country of birth, or income.


# Types of statistics:  
### Descriptive statistics : Consist of organizing and summarizing data.
### Inferential statistics : Consist of data you have measured to form conclusion

## Descriptive statistics : Descriptive statistics refers to the collection, representation, and formation of data.

## There are 3 main types of descriptive statistics:
#### 1-	Measures of central tendency: 
Help you find the middle, or the average, of a dataset.

   - The 3 most common measures of central tendency are:  A- Mode    B- Median   C- Mean.

#### 2-Measures of dispersion: 
Dispersion is the state of getting dispersed or spread. Statistical dispersion means the extent to which numerical data is likely to vary about an average value.

## There are two main types of dispersion methods: 

#### A-	Absolute Measure of Dispersion :

- Absolute Measure of Dispersion: An absolute measure of dispersion contains the same unit as the original data set. The absolute dispersion method expresses the variations in terms of the average of deviations of observations like standard or means deviations.

#### The types of absolute measures of dispersion are: 

- A-Range : It is simply the difference between the maximum value and the minimum value given in a data set.

- B-Variance : Deduct the mean from each data in the set, square each of them and add each square and finally divide them by the total no of values in the data set to get the variance.

- C- Standard Deviation : The square root of the variance is known as the standard deviation

#### B-	Relative Measure of Dispersion :
- The relative measures of dispersion are used to compare the distribution of two or more data sets.


#### 3-Different type of distribution of data :
- Histogram 
- PDF Probability Distribution Functions : The PDF is a function that describes the probability of a continuous random variable taking on a certain value. 
- PMF Probability mass function : The PMF is a function that describes the probability of a discrete random variable taking on a certain value. 
- CDF Cumulative distribution function : The CDF is a function that describes the probability that a random variable (continuous or discrete) will take on a value less than or equal to a certain value.

## Inferential statistics :  Inferential statistics use a random sample to draw conclusions about the population.

-In contrast to descriptive statistics, inferential statistics want to make a statement about the population.


## What are the 3 types of inferential statistics?

- Hypothesis testing.
- Confidence intervals. 
- Regression analysis. 

- t-tests : A t-test is a statistical test that can be used to compare means
- Z-test : A z-test is a statistical test used to determine whether two population means are different when the variances are known and the sample size is large.
- ANOVA (Analysis of Variance) : An ANOVA is a statistical test that is also used to compare means. 
- F-test : is a parametric test that helps the researcher draw out an inference about the data that is drawn from a particular population.
- Chi-Square :  is one of the inferential statistics that is used to formulate and check the interdependence of two or more variables

## Does central tendency mean average?
- Measures of central tendency help you find the " middle, or the average " , of a data set. 
- The 3 most common measures of central tendency are the mean, median and mode. The mode is the most frequent value. The median is the middle number in an ordered data set.

## Population and Sample data :

## population : 
- A population is the entire group that you want to draw conclusions about.

## Sample data : 
- A sample is the specific group that you will collect data from. The size of the sample is always less than the total size of the population.

## Population and Sample Examples
- All the students in the class are population whereas the top 10 students in the class are the sample
